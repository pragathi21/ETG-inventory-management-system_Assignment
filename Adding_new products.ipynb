
Read data from JSON

```python
fd=open("bakery.json",'r')
r=fd.read()
fd.close()
```

```python
r
```

```python
import json
```

```python
bakery=json.loads(r)
```

```python
 bakery
```

add new item into inventory

```python
prod_id=str(input(" enter product id:"))
name=str(input(" enter name:"))
price=int(input("enter price:"))
quantity=int(input("enter quantity:"))
discount=int(input("enter discount:"))
expiry_date=int(input("enter expiry_date"))
bakery[prod_id]={'name':name,'price':price,'quantity':quantity,'discount':discount,'expiry_date:':expiry_date}
js=json.dumps(bakery)

fd=open("bakery.json",'w')
fd.write(js)
fd.close()

```
